```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('../images/green_parrot.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(20, 10))
plt.imshow(image);

plt.show()

image = cv2.cvtColor(image, cv2.COLOR_RGB2Lab)
```

```python
lightness = image[:,:,0]
a = image[:,:,1]
b = image[:,:,2]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,5))

ax[0].hist(lightness.ravel(), bins = 256);
ax[0].set_title('Distribution of Lightness value')
ax[0].set_xlabel('Pixel value')

ax[1].hist(a.ravel(), bins = 256);
ax[1].set_title('A value (big values - more red, low values - more green)')
ax[1].set_xlabel('Pixel value')

ax[2].hist(b.ravel(), bins = 256);
ax[2].set_title('B value (big values - more yellow, low values - more blue)')
ax[2].set_xlabel('Pixel value')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,5))

ax[0].imshow(lightness, cmap='gray')
ax[0].axis('off')

ax[1].imshow(a, cmap='gray')
ax[1].axis('off')

ax[2].imshow(b, cmap='gray')
ax[2].axis('off')
```
