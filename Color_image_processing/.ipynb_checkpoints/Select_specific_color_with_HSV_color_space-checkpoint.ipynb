```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import copy
```

```python
image = cv2.imread('../images/green_parrot.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

fig, ax = plt.subplots(1, 2, figsize=(15,5))

ax[0].imshow(image, cmap='gray');
ax[0].axis('off')

ax[1].imshow(image_hsv, cmap='gray');
ax[1].axis('off')

plt.show()
```

```python
hue = image_hsv[:,:,0]
saturation = image_hsv[:,:,1]
intensity = image_hsv[:,:,2]

fig, ax = plt.subplots(1, 3, figsize=(15,5))

ax[0].imshow(hue, cmap='gray')
ax[0].axis('off')

ax[1].imshow(saturation, cmap='gray')
ax[1].axis('off')

ax[2].imshow(intensity, cmap='gray')
ax[2].axis('off')
```

# cv2.inRange()

Checks if array elements lie between the elements of two other arrays.

**Parameters:**

1. src	first input array.
2. lowerb	inclusive lower boundary array or a scalar.
3. upperb	inclusive upper boundary array or a scalar.
4. dst	output array of the same size as src and CV_8U type.


```python
# Select red part of image

upper_red = np.array([10, 255, 255])
lower_red = np.array([0, 255, 255])

mask = cv2.inRange(image_hsv, lower_red, upper_red)

red_result = cv2.bitwise_and(image, image, mask=mask)

fig, ax = plt.subplots(1, 2, figsize=(15,5))

ax[0].imshow(image, cmap='gray')
ax[0].axis('off')

ax[1].imshow(red_result, cmap='gray')
ax[1].axis('off')
```

```python
# Select green and blue parts and make them red

upper_blue = np.array([130, 255, 255])
lower_blue = np.array([110, 255, 255])

upper_green = np.array([85, 255, 255])
lower_green = np.array([50, 255, 255])

mask_blue = cv2.inRange(image_hsv, lower_blue, upper_blue)
mask_green = cv2.inRange(image_hsv, lower_green, upper_green)

red_values = np.unique(red_result[0, :, :])
print(red_values)

mask = mask_blue + mask_green

red_image = copy.deepcopy(image)

red_image[mask>0] = (255, 0, 0)

result = cv2.bitwise_and(image, image, mask=mask)

fig, ax = plt.subplots(1, 2, figsize=(15,5))

ax[0].imshow(red_image, cmap='gray')
ax[0].axis('off')

ax[1].imshow(result, cmap='gray')
ax[1].axis('off')
```

```python
red_image_hsv = cv2.cvtColor(red_image, cv2.COLOR_RGB2HSV)

hue = red_image_hsv[:,:,0]
saturation = red_image_hsv[:,:,1]
intensity = red_image_hsv[:,:,2]

fig, ax = plt.subplots(1, 3, figsize=(15,5))

ax[0].imshow(hue, cmap='gray')
ax[0].axis('off')

ax[1].imshow(saturation, cmap='gray')
ax[1].axis('off')

ax[2].imshow(intensity, cmap='gray')
ax[2].axis('off')
```

```python
# Select green part of image with parrots
image = cv2.imread('../images/green_parrot.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

upper_green = np.array([85, 255, 255])
lower_green = np.array([40, 0, 0])

mask = cv2.inRange(image_hsv, lower_green, upper_green)

red_result = cv2.bitwise_and(image, image, mask=mask)

fig, ax = plt.subplots(1, 2, figsize=(15,5))

ax[0].imshow(image, cmap='gray');
ax[0].axis('off')

ax[1].imshow(red_result, cmap='gray');
ax[1].axis('off')

plt.show()
```
