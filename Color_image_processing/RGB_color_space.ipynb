```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('../images/green_parrot.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(20, 10))
plt.imshow(image);

plt.show()
```

```python
red = image[:,:,0]
green = image[:,:,1]
blue = image[:,:,2]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,5))

ax[0].hist(red.ravel(), bins = 256, color='red');
ax[0].set_title('Distribution of red pixels')
ax[0].set_xlabel('Pixel value')

ax[1].hist(green.ravel(), bins = 256, color='g');
ax[1].set_title('Distribution of green pixels')
ax[1].set_xlabel('Pixel value')

ax[2].hist(blue.ravel(), bins = 256);
ax[2].set_title('Distribution of blue pixels')
ax[2].set_xlabel('Pixel value')

plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15,5))

ax[0].imshow(red, cmap='gray')
ax[0].axis('off')

ax[1].imshow(green, cmap='gray')
ax[1].axis('off')

ax[2].imshow(blue, cmap='gray')
ax[2].axis('off')
```
