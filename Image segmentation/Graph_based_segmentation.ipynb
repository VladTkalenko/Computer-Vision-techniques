```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

# cv2.ximgproc.segmentation.createGraphSegmentation()
Create a graph based segmentor.

**Parameters:**

1. sigma	The sigma parameter, used to smooth image
2. k	The k parameter of the algorythm. Bigger k causes a preference for larger components.
3. min_size	The minimum size of segments

```python
gc = cv2.ximgproc.segmentation.createGraphSegmentation(sigma=1.5, k=300, min_size=50)
```

```python
gc
```

```python
image = cv2.imread('../images/Village.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(20, 10))
plt.imshow(image)

plt.show()
```

```python
out1 = gc.processImage(image[:, :, 0])
out2 = gc.processImage(image[:, :, 1])
out3 = gc.processImage(image[:, :, 2])

final_output = np.zeros((out1.shape[0], out1.shape[1], 3))
final_output[:, :, 0] = out1
final_output[:, :, 1] = out2
final_output[:, :, 2] = out3
img_normalized = cv2.normalize(final_output, None, 0, 1.0,
cv2.NORM_MINMAX, dtype=cv2.CV_32F)
```

```python
fig = plt.figure(figsize=(20, 10))
plt.imshow(img_normalized)
```
