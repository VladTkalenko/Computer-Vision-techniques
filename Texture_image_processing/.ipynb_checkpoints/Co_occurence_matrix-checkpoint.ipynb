```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

import skimage.feature as feature
import skimage.measure as measure
```

```python
test_image = np.array([
    [1, 1, 7, 5, 3, 2],
    [5, 1, 6, 1, 2, 5],
    [8, 8, 6, 8, 1, 2],
    [4, 3, 4, 5, 5, 1],
    [8, 7, 8, 7, 6, 2],
    [7, 8, 6, 2, 6, 2]
])
```

# skimage.feature.graycomatrix

Calculate the gray-level co-occurrence matrix.
A gray level co-occurrence matrix is a histogram of co-occurring grayscale values at a given offset over an image.


**Parameters:**

1. image (array_like) - Integer typed input image. Only positive valued images are supported. If type is other than uint8, the argument levels needs to be set.
2. distances (array_like) - List of pixel pair distance offsets.
3. angles (array_like) -  List of pixel pair angles in radians.
4. levels (int, optional) - The input image should contain integers in [0, levels-1], where levels indicate the number of gray-levels counted (typically 256 for an 8-bit image). This argument is required for 16-bit images or higher and is typically the maximum of the image. As the output matrix is at least levels x levels, it might be preferable to use binning of the input image rather than large values for levels.
5. symmetric (bool, optional) -  If True, the output matrix P[:, :, d, theta] is symmetric. This is accomplished by ignoring the order of value pairs, so both (i, j) and (j, i) are accumulated when (i, j) is encountered for a given offset. The default is False.
6. normed (bool, optional) - If True, normalize each matrix P[:, :, d, theta] by dividing by the total number of accumulated co-occurrences for the given offset. The elements of the resulting matrix sum to 1. The default is False.

**Output:**

P (4-D ndarray) - The gray-level co-occurrence histogram. The value P[i,j,d,theta] is the number of times that gray-level j occurs at a distance d and at an angle theta from gray-level i. If normed is False, the output is of type uint32, otherwise it is float64. The dimensions are: levels x levels x number of distances x number of angles.

```python
gray_co_matrix = feature.graycomatrix(test_image, [1], [0], levels=9)
```

```python
print(gray_co_matrix.shape)

print(np.squeeze(gray_co_matrix))
```

# Skimage.feature.graycoprops

Calculate texture properties of a GLCM.

Compute a feature of a gray level co-occurrence matrix to serve as a compact summary of the matrix. 

Features which can be computed:

1. Contrast - A measure of intensity contrast between a pixel and its neighbor over the entire image. 
$$\text{contrast} = \sum_{i=1}^{K}{\sum_{j=1}^{K}{(i-j)^2*\text{p}_{ij}}}$$
2. Dissimilarity.
$$\text{Dissimilarity} = \sum_{i=1}^{K}{\sum_{j=1}^{K}{\left\lvert i-j \right\rvert}*\text{p}_{ij}}$$
3. Homogeneity - Measures the spatial closeness to the diagonal of the distribution of elements in GLCM. The range of values is [0, 1], with the maximum being achieved when GLCM is a diagonal matrix.
$$\text{Homogeneity} = \sum_{i=1}^{K}{\sum_{j=1}^{K}{\frac{\text{p}_{ij}}{1 + \left\lvert i-j \right\rvert}}}$$
4. ASM - A measure of uniformity in the range [0, 1]. Uniformity is 1 for a constant image.
$$\text{ASM} = \sum_{i=1}^{K}{\sum_{j=1}^{K}{\text{p}_{ij}^2}}$$
5. Energy.
$$\text{Energy} = \sqrt{ASM}$$
6. Correlation - a measure of how correlated a pixel is to its neighbor over the entire image. The range of values is 1 to -1 corresponding to perfect positive and perfect negative correlations. This measure is not defined if either standart deviation is zero.
$$\text{Correlation} = \sum_{i=1}^{K}{\sum_{j=1}^{K}{\frac{\text{p}_{ij}(i-\mu_{i})(j-\mu_{j})}{(\sigma_{i})^2(\sigma_{j})^2}}}$$ 

**Parameters:**

1. P (ndarray) - Input array. P is the gray-level co-occurrence histogram for which to compute the specified property. The value P[i,j,d,theta] is the number of times that gray-level j occurs at a distance d and at an angle theta from gray-level i.
2. prop {‘contrast’, ‘dissimilarity’, ‘homogeneity’, ‘energy’, ‘correlation’, ‘ASM’}, optional - The property of the GLCM to compute. The default is ‘contrast’.

**Output:**

results (2-D ndarray) - 2-dimensional array. results[d, a] is the property ‘prop’ for the d’th distance and the a’th angle.

# skimage.measure.shannon_entropy

Calculate the Shannon entropy of an image.
$$\text{entropy} = -\sum_{i=1}^{K}{\sum_{j=1}^{K}{\log_{base}{\text{p}}_{ij}*\text{p}_{ij}}}$$

**Parameters:**

1. image ((N, M) ndarray) - Grayscale input image.
2. base (float, optional) -  The logarithmic base to use.

**Returns:**

1. entropy (float)

```python
contrast = feature.graycoprops(gray_co_matrix, 'contrast')
dissimilarity = feature.graycoprops(gray_co_matrix, 'dissimilarity')
homogeneity = feature.graycoprops(gray_co_matrix, 'homogeneity')
ASM = feature.graycoprops(gray_co_matrix, 'ASM')
energy = feature.graycoprops(gray_co_matrix, 'energy')
correlation = feature.graycoprops(gray_co_matrix, 'correlation')
entropy = measure.shannon_entropy(gray_co_matrix)
```

```python
print(f"Contrast = {round(contrast[0][0], 2)}")
print(f"Dissimilarity = {round(dissimilarity[0][0], 2)}")
print(f"Homogeneity = {round(homogeneity[0][0], 2)}")
print(f"ASM = {round(ASM[0][0], 2)}")
print(f"Energy = {round(energy[0][0], 2)}")
print(f"Correlation = {round(correlation[0][0], 2)}")
print(f"Entropy = {round(entropy, 2)}")
```

```python
image_flat = cv2.imread('../images/flat.jpg')
image_flat_gray = cv2.cvtColor(image_flat, cv2.COLOR_BGR2GRAY)

image_village = cv2.imread('../images/Village.jpg')
image_village_gray = cv2.cvtColor(image_village, cv2.COLOR_BGR2GRAY)

image_flat = cv2.cvtColor(image_flat, cv2.COLOR_BGR2RGB)
image_village = cv2.cvtColor(image_village, cv2.COLOR_BGR2RGB)

fig, axs = plt.subplots(2, 2, figsize=(20, 20))
axs[0, 0].imshow(image_flat);
axs[0, 0].axis('Off')

axs[0, 1].imshow(image_flat_gray, cmap='gray');
axs[0, 1].axis('Off')

axs[1, 0].imshow(image_village);
axs[1, 0].axis('Off')

axs[1, 1].imshow(image_village_gray, cmap='gray');
axs[1, 1].axis('Off')

plt.show()
```

```python
village_co_matrix = feature.graycomatrix(image_village_gray, [1], [0])
village_co_matrix_norm = np.squeeze(village_co_matrix/np.max(village_co_matrix).astype(int))

flat_co_matrix = feature.graycomatrix(image_flat_gray, [1], [0])
flat_co_matrix_norm = np.squeeze(flat_co_matrix/np.max(flat_co_matrix).astype(int))

fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].imshow(village_co_matrix_norm, vmin=0, vmax=.01)
axs[0].set_title('Co-occurence matrix Village')
axs[0].axis('Off')

axs[1].imshow(flat_co_matrix_norm, vmin=0, vmax=.01)
axs[1].set_title('Co-occurence matrix Flat')
axs[1].axis('Off')
```

```python
village_descriptors = {
    "Contrast": 0,
    "Dissimilarity": 0,
    "Homogeneity": 0,
    "ASM": 0,
    "Energy": 0,
    "Correlation": 0,
    "Entropy": 0
}

flat_descriptors = {
    "Contrast": 0,
    "Dissimilarity": 0,
    "Homogeneity": 0,
    "ASM": 0,
    "Energy": 0,
    "Correlation": 0,
    "Entropy": 0
}

descriptors = [village_descriptors, flat_descriptors]
matrices = [village_co_matrix, flat_co_matrix]
```

```python
for i in range(2):
    descriptors[i]['Contrast'] = feature.graycoprops(matrices[i], 'contrast')[0][0]
    descriptors[i]['Dissimilarity'] = feature.graycoprops(matrices[i], 'dissimilarity')[0][0]
    descriptors[i]['Homogeneity'] = feature.graycoprops(matrices[i], 'homogeneity')[0][0]
    descriptors[i]['ASM'] = feature.graycoprops(matrices[i], 'ASM')[0][0]
    descriptors[i]['Energy'] = feature.graycoprops(matrices[i], 'energy')[0][0]
    descriptors[i]['Correlation'] = feature.graycoprops(matrices[i], 'correlation')[0][0]
    descriptors[i]['Entropy'] = measure.shannon_entropy(matrices[i])
```

```python
keys = list(descriptors[0].keys())
for key in keys:
    print(f"{key} in village photo = {round(descriptors[0][key], 4)}")
    print(f"{key} in flat photo = {round(descriptors[1][key], 4)}")
    print(' ')
```

```python
distances = np.linspace(1, 50, 50).astype(int)

co_matrix_flat = feature.graycomatrix(image_flat_gray, distances, [0])
co_matrix_village = feature.graycomatrix(image_village_gray, distances, [0])

corelations_flat = feature.graycoprops(co_matrix_flat, 'correlation')
corelations_village = feature.graycoprops(co_matrix_village, 'correlation')
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(range(len(distances)), np.squeeze(corelations_flat.T))
ax[0].set_xlabel('Distance between pixels')
ax[0].set_ylabel('Correlation')
ax[0].set_title('Correlation of pixels in flat image')

ax[1].plot(range(len(distances)), np.squeeze(corelations_village.T))
ax[1].set_xlabel('Distance between pixels')
ax[1].set_ylabel('Correlation')
ax[1].set_title('Correlation of pixels in village image')
```
